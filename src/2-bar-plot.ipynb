### Exercise 2: Bar Plot with Colors

```python
import matplotlib.pyplot as plt

# Data
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]

# Colors for each bar
colors = ['red', 'green', 'blue', 'orange']

# Create the bar plot
plt.bar(categories, values, color=colors)

# Add a title
plt.title('Bar Plot of Categories', fontsize=14)

# Show the plot
plt.show()
```
