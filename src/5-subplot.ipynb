### Exercise 5: Subplot Layout

```python
import matplotlib.pyplot as plt
import numpy as np

# Create a 2x2 subplot layout
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Data for the plots
x = np.linspace(0, 10, 100)
y_line = x ** 2
categories = ['A', 'B', 'C', 'D']
values_bar = [5, 7, 3, 9]
x_scatter = [1, 2, 3, 4, 5]
y_scatter = [2, 4, 6, 8, 10]
labels_pie = ['Python', 'Java', 'C++', 'JavaScript']
sizes_pie = [40, 25, 20, 15]

# First subplot: Line chart
axs[0, 0].plot(x, y_line)
axs[0, 0].set_title('Line Chart')

# Second subplot: Bar chart
axs[0, 1].bar(categories, values_bar, color=['red', 'green', 'blue', 'orange'])
axs[0, 1].set_title('Bar Chart')

# Third subplot: Scatter plot
axs[1, 0].scatter(x_scatter, y_scatter)
for i in range(len(x_scatter)):
    axs[1, 0].annotate(f'({x_scatter[i]}, {y_scatter[i]})', (x_scatter[i], y_scatter[i]), textcoords="offset points", xytext=(0,5), ha='center')
axs[1, 0].set_title('Scatter Plot')

# Fourth subplot: Pie chart
axs[1, 1].pie(sizes_pie, labels=labels_pie, autopct='%1.1f%%', startangle=90)
axs[1, 1].set_title('Pie Chart')

# Adjust layout for better spacing
plt.tight_layout()

# Show the plots
plt.show()
```
