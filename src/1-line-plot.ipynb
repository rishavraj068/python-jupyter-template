### Exercise 1: Line Plot Customization

```python

import matplotlib.pyplot as plt
import numpy as np

# Data
x = np.linspace(0, 10, 100)
y = x ** 2

# Create the plot
plt.plot(x, y)

# Customize the plot
plt.title('Square of x (x^2)', fontsize=14)
plt.xlabel('x', fontsize=12)
plt.ylabel('y = x^2', fontsize=12)
plt.grid(True)

# Show the plot
plt.show()
```
