### Exercise 3: Scatter Plot with Annotations

```python
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Create the scatter plot
plt.scatter(x, y)

# Annotate each point
for i in range(len(x)):
    plt.annotate(f'({x[i]}, {y[i]})', (x[i], y[i]), textcoords="offset points", xytext=(0,5), ha='center')

# Add title
plt.title('Scatter Plot with Annotations', fontsize=14)

# Show the plot
plt.show()
```
